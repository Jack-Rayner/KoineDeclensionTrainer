```python
def jumpingOnClouds(c):
    current_position = 0
    path = 0
    while current_position <= len(c)-2:
        try:
            if c[current_position + 2] != 1:
                path += 1
            elif c[current_position + 1] != 1:
                path += 1
            else:
            current_position +=1
        except:
            return path+1
    return path
```

```python
jumpingOnClouds([0, 0 ,1, 0, 0, 1, 0])
```
